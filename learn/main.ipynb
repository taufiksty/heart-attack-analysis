### Transform Column to Feature Column

```python
feature_columns = []
```

#### Numeric Column

```python
for header in ['age', 'trtbps', 'chol', 'thalachh', 'oldpeak', 'caa']:
    feature_columns.append(tf.feature_column.numeric_column(header))
feature_columns
```

### Bucketized Column

```python
age = tf.feature_column.numeric_column('age')
age_buckets = tf.feature_column.bucketized_column(age, boundaries=[18, 25, 30, 35, 40, 45, 50, 55, 60, 65])
print(age_buckets)
feature_columns.append(age_buckets)
```

```python
feature_columns
```

### Indicator Columns

```python
df["thall"] = df["thall"].apply(str)
thal = tf.feature_column.categorical_column_with_vocabulary_list(
    "thall", ["3", "6", "7"]
)
thal_one_hot = tf.feature_column.indicator_column(thal)
print(thal_one_hot)
feature_columns.append(thal_one_hot)
```

```python
df["sex"] = df["sex"].apply(str)
sex = tf.feature_column.categorical_column_with_vocabulary_list(
    "sex", ["0", "1"]
)
sex_one_hot = tf.feature_column.indicator_column(sex)
print(sex_one_hot)
feature_columns.append(sex_one_hot)
```

```python
df["cp"] = df["cp"].apply(str)
cp = tf.feature_column.categorical_column_with_vocabulary_list(
    "cp", ["0", "1", "2", '3']
)
cp_one_hot = tf.feature_column.indicator_column(cp)
print(cp_one_hot)
feature_columns.append(cp_one_hot)
```

```python
df["slp"] = df["slp"].apply(str)
slp = tf.feature_column.categorical_column_with_vocabulary_list(
    "slp", ["0", "1", "2"]
)
slp_one_hot = tf.feature_column.indicator_column(slp)
print(slp_one_hot)
feature_columns.append(slp_one_hot)
```

### Embedding Columns

```python
thal_embedding = tf.feature_column.embedding_column(thal, dimension=8)
feature_columns.append(thal_embedding)
print(thal_embedding)
```

```python
feature_columns
```

### Crossed Columns

```python
age_thal_crossed = tf.feature_column.crossed_column(
    [age_buckets, thal], hash_bucket_size=1000
)
age_thal_crossed = tf.feature_column.indicator_column(age_thal_crossed)
print(age_thal_crossed)
feature_columns.append(age_thal_crossed)
```

```python
cp_slp_crossed = tf.feature_column.crossed_column(
    [cp, slp], hash_bucket_size=1000
)
cp_slp_crossed = tf.feature_column.indicator_column(cp_slp_crossed)
print(cp_slp_crossed)
feature_columns.append(cp_slp_crossed)
```

```python
feature_columns
```

### Create Input Pipeline

```python
from sklearn.model_selection import train_test_split
```

```python
def df_to_dataset(df, batch_size=32):
    df = df.copy()
    labels = df.pop("output")

    return (
        tf.data.Dataset.from_tensor_slices((dict(df), labels))
        .shuffle(buffer_size=len(df))
        .batch(batch_size)
    )
```

### Split Data

```python
train, test = train_test_split(df, test_size=.2, random_state=32)

train_ds = df_to_dataset(train)
test_ds = df_to_dataset(test)
```

```python
print(test_ds)
```

### Preview Input Pipeline

```python
for feature_batch, label_batch in train_ds.take(2):
    print('Every feature:', list(feature_batch.keys()))
    print('A batch of ages:', feature_batch['age'])
    print('A batch of outputs:', label_batch)
```

### Feature Layer

```python
feature_layer = tf.keras.layers.DenseFeatures(feature_columns)
feature_layer
```

### Model

```python
tf.keras.backend.set_floatx('float64')

model = tf.keras.Sequential([
    feature_layer,
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dropout(rate=0.2),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])
```

### Compile

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

### Train

```python
history = model.fit(train_ds, 
                    validation_data=test_ds,
                    epochs=100,
                    use_multiprocessing=True)
```

### Evaluate Model

```python
model.evaluate(test_ds)
```

### Plot Model Accuracy

```python
plt.figure(figsize=(10, 6))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim((0, 1))
plt.legend(['Train', 'Test'], loc='upper left')
```

### Plot Model Loss

```python
plt.figure(figsize=(10, 6))
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(["Train", "Test"], loc="upper left")
```

### Predict Test Dataset

```python
print(test_ds)
predictions = model.predict(test_ds)
bin_predictions = tf.round(predictions).numpy().flatten()
bin_predictions
```
